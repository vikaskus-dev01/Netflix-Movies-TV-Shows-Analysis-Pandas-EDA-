```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r"C:/Users/kushw/Downloads/netflix_titles.csv.zip")
df.head()
```

```python
df.info()
df.describe(include='all')
df.isnull().sum()
df.columns
df.head()
```

```python
df['director'].fillna("Unknown", inplace=True)
df['cast'].fillna("Unknown", inplace=True)
df['country'].fillna("Unknown", inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
```

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV shows")
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
plt.title("Top content Producing countries")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['listed_in'].value_counts().head(10).plot(kind='bar')
plt.title("Top Genres on Netflix")
plt.show()
```

```python
df['year_added'].value_counts().sort_index().plot(kind='bar')
plt.title("Netflix Additions by Year")
plt.show()
```

```python
df['director'].value_counts().head(10).plot(kind='bar')
plt.title("Most Content Created by Directors")
plt.show()
```
